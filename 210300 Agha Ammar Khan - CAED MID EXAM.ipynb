```python
def grades(dataframe):
    grades=np.array([])
    grandTotalColumn=np.array(dataframe)
    for i in grandTotalColumn:
        if i >= 85:
            grades=np.append(grades,"A")
        elif i>=75:
             grades=np.append(grades,"B+")
        elif i>=70:
             grades=np.append(grades,"B")
        elif i>=65:
             grades=np.append(grades,"C+")
        elif i>=60:
             grades=np.append(grades,"C")
        elif i>=50:
             grades=np.append(grades,"D")
        elif i<50:
             grades=np.append(grades,"F")
            
    grandTotalColumn=pd.DataFrame(grandTotalColumn)
    grades=pd.DataFrame(grades)
    grading=pd.concat([grades],axis=1)
    grading=np.array(grading)
    grading=pd.DataFrame(grading)
    return grading
```

```python
def Result(dictionary,pandaDataFrame):
    df=pd.Series(dictionary)
    array=np.array(pandaDataFrame)
    LR=array[:,2:16]
    LP=array[:,16:30]
    MT=array[:,30]
    FT=array[:,31]
    CEA=array[:,32]
    rollNumber=array[:,0]
    Name=array[:,1]
    LP=np.sum(LP,axis=1)
    LR=np.sum(LR,axis=1)
    LRTotal=np.multiply((df['LR']),(LR/210))
    LPTotal=np.multiply((df['LP']),(LP/210))
    MTTotal=np.multiply((df['MT']),(MT/55))
    FTTotal=np.multiply((df['FT']),(FT/50))
    CEATotal=np.multiply((df["CEA"]),(CEA/20))
    rollNumber=pd.DataFrame(rollNumber)
    Name=pd.DataFrame(Name)
    lr=pd.DataFrame(LRTotal)
    lp=pd.DataFrame(LPTotal)
    mt=pd.DataFrame(MTTotal)
    ft=pd.DataFrame(FTTotal)
    cea=pd.DataFrame(CEATotal)
    result=pd.concat([lp,lr,mt,ft,cea],axis=1)
    GrandTotal=np.sum(result,axis=1)
    GrandTotal=np.array(GrandTotal)
    GrandTotal=pd.DataFrame(GrandTotal)
    Grade=grades(GrandTotal)
    result=pd.concat([rollNumber,Name,lp,lr,mt,ft,cea,GrandTotal,Grade],axis=1)
    result=np.array(result)
    result=pd.DataFrame(result)
    result=np.array(result)
    newFile=pd.DataFrame(result,columns=["Roll numbers","Names","Lab Performance Total","Lab Report Total","Mid Term Total","Final Term Total","CEA Total","Grand Total","Grades"])
    return newFile
   
```
